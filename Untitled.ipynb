```python
print("Welcome to  Grace Pizza")
print("here you will find what ever yo desire for")
print("currently we have two categories")
print("pizzas and burgers")
zinger_burger = 16.99;
zinger_tower = 5.99;
crispy_burger = 19.99;
special_burger = 10.50;
pizza_tower = 19.01;
cheese_pizza = 18.99;
fajita_pizza = 23.7;
pizza_masala = 12.88;
special_fajita_pizza = 13.99;
income_tax = 1.2;
total = 0.0;
h=str(input("enter the category for menu: "))
if h == "Burgers":
    print('''1.special burger
2.zinger Burger
3.Zinger Tower
4.crispy burger''')
elif h == "pizzas":
    print('''1.pizza tower 
2.cheese pizza
3.fajita Pizza
4.pizza masala
5.special Fajita pizza''')
else:
    print("invalid category Input")
complete=True
order=str(input("enter the number of  things you want or press q for quit"))
while (complete):
    print("Total amount: ", total)
    if order == "1":
        total += zinger_burger;
    elif order == "2":
        total += zinger_tower;
    elif order == "3":
        total += crispy_burger;
    elif order == "4":
        total += fajita_pizza;
    elif order == "5":
        total += special_fajita_pizza;
    elif order == "q":
        print("Final total:", total*income_tax);
        complete= False  
    
    print("Final total:", total*income_tax);
    print("Thanks for buying meal from us.")
    print("stay home stay safe")
```
